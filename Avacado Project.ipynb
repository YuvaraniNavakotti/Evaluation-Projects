### Model Building 

### Predicting Average Price - Regression Problem

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

Scaling the feature variable using standard scalar.

```python
X_sc=StandardScaler().fit_transform(X)
```

##### Finding best Randomstate

```python
m_acc=0
b_rs=0
for i in range(1,200):
    lr=LinearRegression()
    X_train,X_test,y_train,y_test=train_test_split(X_sc,y,test_size=0.25,random_state=i)
    lr.fit(X_train,y_train)
    y_pred=lr.predict(X_test)
    acc=r2_score(y_test,y_pred)
    if(acc>m_acc):
        m_acc=acc
        b_rs=i
print("Best Accuracy: ",m_acc,"Best RS: ",b_rs)
```

Splitting the train test data set using best random state found.

```python
X_train,X_test,y_train,y_test=train_test_split(X_sc,y,test_size=0.25,random_state=55)
```

```python
lr=LinearRegression()
rf_reg=RandomForestRegressor()
knn_reg=KNeighborsRegressor()
ls=Lasso()
rd_reg=Ridge()
svm_reg=SVR()

models=[lr,rf_reg,knn_reg,ls,rd_reg,svm_reg]
```

#### Building base models - Model selection

```python
for m in models:
    m.fit(X_train,y_train)
    y_pred=m.predict(X_test)
    print("Metrics for ",m)
    print("Trian Accuracy: ",m.score(X_train,y_train))
    print("Test Accuracy: ",r2_score(y_test,y_pred))
    print("RMSE: ",np.sqrt(mean_squared_error(y_test,y_pred)))
```

#### Cross Validation check

```python
for m in models:
    print("Cross Validation score for ",m, cross_val_score(m,X_sc,y,cv=5).mean())
```

##### Based on RMSE score and the Cross validation score RandomForestregressor works fine for our dataset.

```python
m_acc=0
b_rs=0
for i in range(1,10):
    model=RandomForestClassifier(n_jobs=-1,criterion='entropy',max_features='log2', n_estimators=150,random_state=i)
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    acc=r2_score(y_test,y_pred)
    if(acc>m_acc):
        m_acc=acc
        b_rs=i
print("Best Accuracy: ",m_acc,"Best RS: ",b_rs)
model=RandomForestClassifier(n_jobs=-1,criterion='entropy',max_features='log2',n_estimators=150,random_state=b_rs)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print("Metrics for ",model)
print("Trian Accuracy: ",model.score(X_train,y_train))
print("Test Accuracy: ",accuracy_score(y_test,y_pred))
print("Confusion Matrix:\n ",confusion_matrix(y_test,y_pred))
print('Classification report:\n',classification_report(y_test,y_pred))
```
